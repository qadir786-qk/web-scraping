```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python


                                                # WEB SCRAPPING

```

```python
web = requests.get("https://quotes.toscrape.com/")
print(web)
```

```python
soup = BeautifulSoup(web.content, 'html.parser')
print(soup.prettify())
```

```python
# EXTRACTING QUOTES

class2 = soup.find_all("span", class_="text")

quotes = [tag.get_text(strip=True) for tag in class2]
print(quotes)
```

```python
# EXTRACTING AUTHORS' NAMES

class3 = soup.find_all("small", class_="author")

authors = [tag.get_text(strip=True) for tag in class3]
print(authors)
```

```python
# CREATING A DATAFRAME(TABLE)

df2 = pd.DataFrame({"authors": authors, "quotes": quotes})
print(df2)

df2.to_csv("authors' quotes.csv", index=False)
```
